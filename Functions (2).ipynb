## 1)global variable:

## 2)local variables:

```python
The variables which are declared Inside the function called as local variables
Once we create the local variable we can't use those variables outside the function.
used within the same function
syntax:
    def f1():
        local variable
        ---------------
        ---------------      
    def f2():
        local variable
        --------------
        --------------
```

```python
#eg pgm for local variable
def fun1():
    x=10 #local variable
    print("a is local variable to fun1",x)
def fun2():
    print("printing fun1 local variable in fun2",x)
fun1()
fun2()
```

```python
#eg pgm for local variable
def fun1():
    global y #while declaring variables we can't initialize values when converting local to global
    y=10
    print("a is local variable to fun1",y)
def fun2():
    print("printing fun1 local variable in fun2",y)
fun1()
fun2()
```

# Anonymous function

```python
functions without names i.e, nameless functions are anonymous functions
advantage: for instant usage we use anonymous function
mostly we use in ethical hacking and data science
We use the keyword "lambda"

```

## lambda function

```python
lambda: for instant usage and no need to call the function
No Return statement in lambda
To pass the function as an argument to other function also we use lambda function
syntax:
    lambda argumentlist:expression
```

```python
#pgm to perform square of given number
def sqr(n):
    return n*n
x=sqr(2)
print(x)
```

```python
#pgm to perform square of given number using anonymous function
s=lambda n:n*n
print("square of 4 is:",s(4))
```

```python
#pgm to find the sum of 2 no's
x=lambda a,b:a+b
print("sum of 10,20 is:",x(10,20))
```

```python
#pgm to find biggest of 2 no's
y=lambda a,b:a if a>b else b
print("biggest num is:",y(30,20))
```

### filter():

```python
We can use filter() to filter the values from the given sequence based on some conditions.

syntax:
    
    filter(function,sequence)
```

```python
#pgm to filter even numbers in the given list
l=[1,3,4,2,5,6,8,20,13,15]
l1=list(filter(lambda x:x%2==0,l))
print(l1)
```

```python
#pgm to filter given no's in the given list
l=[1,3,4,2,5,6,8,20,13,15]
l1=list(filter(lambda x:x%2!=0,l))
print(l1)
```

```python
#pgm to print even no's of list using functions without filter function
def even(x):
    if x%2==0:
        return True
    else:
        return False
l=[1,3,4,2,5,6,8,20,13,15]
l1=list(filter(even,l))
print(l1)
```

### map()

```python
For every element present in the sequence if we want to apply any functionality like multiplication then map function is used
```

```python
#pgm to multiply each element in list with 2 without lambda function
l=[1,2,3,4,5,6]
def mult(x):
    return x*2
l1=list(map(mult,l))
print(l1)
```

```python
#pgm to multiply each element in list with 2 with lambda function
l=[1,2,3,4,5,6]
l1=list(map(lambda x:x*2,l))
print(l1)
```

### reduce()

```python

This function is used to convert sequence of elements to a single element
We should import reduce() from functools
```

```python
from functools import *
l=[1,2,3,4,5,6]
s=reduce(lambda x,y:x+y,l)
print(s)
```
